```python
import pandas as pd
df = pd.read_csv('C:/Users/10920/Desktop/代码/iris/iris.csv', header=0 ,engine='python')
import matplotlib.pyplot as plt

# 第一个图
plt.scatter(df.loc[df.Name == 'Iris-setosa', 'PetalLength'], df.loc[df.Name == 'Iris-setosa','PetalWidth'], c='b', label='Iris-setosa')
plt.scatter(df.loc[df.Name == 'Iris-versicolor', 'PetalLength'], df.loc[df.Name == 'Iris-versicolor','PetalWidth'], c='yellow', label='Iris-versicolor')
plt.scatter(df.loc[df.Name == 'Iris-virginica', 'PetalLength'], df.loc[df.Name == 'Iris-virginica','PetalWidth'], c='saddlebrown', label='Iris-virginica')
# 添加轴标签
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
# 添加图例
plt.legend(loc = 'upper left')
plt.show()
```

```python
# 第二个图
plt.scatter(df.loc[df.Name == 'Iris-setosa', 'SepalLength'], df.loc[df.Name == 'Iris-setosa','SepalWidth'], c='b', label='Iris-setosa')
plt.scatter(df.loc[df.Name == 'Iris-versicolor', 'SepalLength'], df.loc[df.Name == 'Iris-versicolor','SepalWidth'], c='yellow', label='Iris-versicolor')
plt.scatter(df.loc[df.Name == 'Iris-virginica', 'SepalLength'], df.loc[df.Name == 'Iris-virginica','SepalWidth'], c='saddlebrown', label='Iris-virginica')
# 添加轴标签
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
# 添加图例
plt.legend(loc = 'upper left')
plt.show()
```
