```python
"""
任务：
1. 忽略数据标准化,使用Kmeans聚类,设定k值为3,输出所有样本点的聚类结果,3个类别的类均值
2. 尝试以花萼长度(sepal length),花瓣长度(petal length)为x,y轴,可视化聚类结果
"""
```

```python
import pandas as pd
from sklearn import datasets
df = pd.DataFrame(datasets.load_iris()['data'], columns=datasets.load_iris()['feature_names'])
```

```python
# 1. 建立聚类模型
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3).fit(df)
# 预测结果
y = kmeans.predict(df)
print(y)
# 3个类均值
print(kmeans.cluster_centers_)
```

```python
# 2. 可视化聚类结果
import seaborn as sns
sns.scatterplot(x=df.loc[:, "sepal length (cm)"], y=df.loc[:, "petal length (cm)"], hue=y)
```
