```python
import pandas as pd
```

```python
login_data = pd.read_csv('C:/Users/10920/Desktop/login_data.csv')
login_data
```

### 需求1：创建一列，根据国内的法定假期规定，从登录时间确定当天是否是工作日

```python
# 数据集中的日期是从2020/7/1至2020/7/31，其中只有周末是休息日，无假期日
# 转成datetime格式
login_data['login_at'] = pd.to_datetime(login_data['login_at'])
# 星期一到星期天为 0-6
login_data['dayofweek'] = login_data['login_at'].dt.dayofweek 
# 创建新列is_work_day，1表示工作日，0表示休息日
login_data['is_work_day'] = login_data.dayofweek.apply(lambda x: 0 if x==6 or x==5 else 1)
login_data
```

### 需求2：统计节假日的平均每日独立登录用户数，工作日的平均每日独立登录用户数

```python
login_data['login_at'] = login_data['login_at'].dt.date
```

```python
login_date1 = login_data.groupby('login_at').agg({'id':pd.Series.nunique, 'is_work_day':'mean'}).rename(columns={"id": "id_cnt"})
login_date_res = login_date1.groupby('is_work_day').agg({'id_cnt':'mean'}).rename(columns={"id_cnt": "id_cnt_avg"})
```

#### 从结果中可以看出，节假日的平均每日独立登录用户数为136.375，工作日的平均每日独立登录用户数为137.478261
